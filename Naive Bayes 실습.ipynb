```python
text = [("I love you", "P"),
        ("love happy weekend", "P"),
        ("bore work job", "N"),
        ("I hate you", "N"),
        ("bore weekend", "N"),
        ("happy together", "P")]

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
token = []
for i in text:
  token.append(i[0].split(' '))
token
```

```python
vector = CountVectorizer(tokenizer = lambda x: x, lowercase=False)
m = vector.fit_transform(token)
X = m.toarray().tolist()
X
```

```python
PN = []
for i in text:
  PN.append(i[1])
PN
```

```python
y = []
for i in PN:
  if i == 'P':
    y.append(1)
  else:
    y.append(0)
y
```

```python
clf = MultinomialNB(force_alpha = True)
clf.fit(X, y)
```

# 예시코드

```python
text = [("I love you", "P"),
        ("love happy weekend", "P"),
        ("bore work job", "N"),
        ("I hate you", "N"),
        ("bore weekend", "N"),
        ("happy together", "P")]
```

```python
X_train = [t[0] for t in text]
Y_train = [t[1] for t in text]
X_train
```

```python
Y_train
```

```python
count_vect = CountVectorizer() # CounterVectorizer 선언
X_train_counts = count_vect.fit_transform(X_train) # fit : 학습
model = MultinomialNB() # MultinomialNB 선언
model.fit(X_train_counts, Y_train)
```

```python
# label 예측
model.predict(count_vect.transform(["happy weekend"]))
```

```python
# label 예측의 확률값
model.predict_proba(count_vect.transform(["happy weekend"]))
```
