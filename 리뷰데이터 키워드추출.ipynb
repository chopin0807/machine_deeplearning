```python
text = list(df['document'])
text
```

```python
# Okt
from konlpy.tag import Okt
okt = Okt()
doc_list = []
for i in text:
  okt_tokens = okt.morphs(i)
  doc_list.append(okt_tokens)
doc_list
```

```python
# 품사부착
TagList = []
for sentence in doc_list:
  oktTag = []
  for token in sentence:
    oktTag += okt.pos(token)
  TagList.append(oktTag)
TagList
```

```python
# 불용어 처리
stopPos = ['Josa', 'Punctuation', 'Suffix', 'Foreign', 'Alpha', 'Number'] # 무조건 제거해야하는 품사
stopWord = ['한', '다', '는', '가', '도', '로'] # 품사가 'Verb'이거나 'Adverb'임, '이', '은', '을', '를'은 품사가 'Noun'임
stopNoun = ['이', '은', '을', '를'] # '이', '은', '을', '를'은 품사가 'Noun'임
doc_word = []
for tag in TagList:
  word = []
  for i in tag:
    if i[1] in stopPos: # 무조건 제외
      continue
    elif i[0] in stopWord and (i[1] == 'Verb' or i[1] == 'Adverb'):
      continue
    elif i[0] in stopNoun and (i[1] == 'Noun'):
      continue
    else:
      word.append(i[0])
  if word == []:
    word.append(i[0])
  doc_word.append(word)
```

```python
doc_word
```

```python
content_token = []
for i in doc_word:
  temp = ' '.join(i)
  content_token.append(temp)
content_token
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer()
tfidf_v = tfidf_vect.fit_transform(content_token)
keyword = tfidf_v.tocoo()
```

```python
sorted_words = sorted(zip(keyword.col, keyword.data), key=lambda x:(x[1], x[0]), reverse=True)
sorted_words
```

```python
feature_name = tfidf_vect.get_feature_names_out()
feature_name
```

```python
# 핵심키워드 10개 추출
[(feature_name[i], score) for i, score in sorted_words[:10]]
```
