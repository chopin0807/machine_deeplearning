```python
training_set = [['me free lottery', 1], ['free get free you', 1], ['you free scholarship', 0], ['free to contact me', 0], 
               ['you won award', 0], ['you ticket lottery', 1]]
test_set = ['free lottery']
```

```python
# spam인 경우에 대한 단어 쪼개기(split)
spam_words = []
for i in training_set:
    if i[1] == 1:
        spam_words.extend(i[0].split(" "))
print(spam_words)
```

```python
# normal인 경우에 대한 단어 쪼개기(split)
normal_words = []
for i in training_set:
    if i[1] == 0:
        normal_words.extend(i[0].split(" "))
print(normal_words)
```

```python
# spam과 normal에서 쪼개진 모든 단어를 리스트에 담은 후 중복 제거
return_words = []
for i in training_set:
    return_words.extend(i[0].split(" "))
words = set(return_words)
words = list(words)
print(words)
```

```python
# 각 단어에 대한 spam, normal에 대한 dataframe 생성
import pandas as pd
df = pd.DataFrame({"tokens": words, "spam": [0 for i in range(len(words))], "normal": [0 for i in range(len(words))]})
df
```

```python
# 조건에 따라 spam, normal개수 count
for i in spam_words:
    df.loc[df["tokens"] == i, "spam"] += 1
for j in normal_words:
    df.loc[df["tokens"] == j, "normal"] += 1
df
```

```python
# 각각의 조건부 확률 계산
k = 0.5
df["P(w|spam)"] = (k + df["spam"])/ (k * 2 + df["spam"].sum())
df
```

```python
df["P(w|normal)"] = (k + df["normal"]) / (k * 2 + df["normal"].sum())
df
```

```python
import numpy as np
log_df = np.log(df["P(w|spam)"])
df["Log(P(w|spam))"] = log_df
df
```

```python
log_df = np.log(df["P(w|normal)"])
df["Log(P(w|normal))"] = log_df
df
```

```python
total_df = pd.DataFrame({"tokens": ["spam", "normal"], "P(label)": [df["spam"].sum() / (df["spam"].sum() + df["normal"].sum()), 
                                                                    df["normal"].sum() / (df["spam"].sum() + df["normal"].sum())]})
total_df
```

```python
total_log = np.log(total_df["P(label)"])
total_df["Log(label)"] = total_log
total_df
```

```python
# 키워드 생성후 검증
key_word = "free lottery"
key_word_split = key_word.split(" ")
key_word_split
```

```python
# 특정 키워드의 Log(P(w|spam))값 조회
list(df[df["tokens"] == "free"]["Log(P(w|spam))"])
```

```python
# 특정 키워드가 tokens에 없는 경우
list(df[df["tokens"] == "like"]["Log(P(w|spam))"])
```

```python
# key_word가 spam일 확률(log_spam값이 클수록 확률이 높다는 뜻)
log_spam = total_df[total_df["tokens"] == "spam"]["Log(label)"]
for i in key_word_split:
    if list(df[df["tokens"] == i]["Log(P(w|spam))"]) != []:
        log_spam += list(df[df["tokens"] == i]["Log(P(w|spam))"])[0]
log_spam
```

```python
# key_word가 spam이 아니고, normal일 확률(log_normal값이 클수록 확률이 높다는 뜻)
log_normal = total_df[total_df["tokens"] == "normal"]["Log(label)"]
for i in key_word_split:
    if list(df[df["tokens"] == i]["Log(P(w|normal))"]) != []:
        log_normal += list(df[df["tokens"] == i]["Log(P(w|normal))"])[0]
log_normal
```
