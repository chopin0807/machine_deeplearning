```python
docs = ['오늘 동물원에서 원숭이와 코끼리를 봤어',
        '동물원에서 원숭이에게 바나나를 줬어 바나나를']
docs
```

```python
doc_ls = []
for doc in docs:
  doc_ls.append(doc.split(' '))
doc_ls
```

```python
from collections import defaultdict

word2id = defaultdict(lambda : len(word2id))
word2id
```

```python
for doc in doc_ls:
  for token in doc:
    word2id[token]
print(doc)
print(word2id)
```

```python
import numpy as np

BoW_ls = []
for i, doc in enumerate(doc_ls):
  bow = np.zeros(len(word2id), dtype=int)
  for token in doc:
    bow[word2id[token]] += 1
  BoW_ls.append(bow.tolist())
```

```python
BoW_ls
```

```python
# TF 구하기
print(len(doc_ls[0]))
print(len(doc_ls[1]))
```

```python
TF_ls = []
for i in BoW_ls:
  temp = []
  for j in i:
    temp.append(j / 5)
  TF_ls.append(temp)
TF_ls
```

```python
# IDF 구하기
doc_num = 2
doc_word = np.zeros(len(word2id), dtype=int).tolist()
for i, doc in enumerate(doc_ls):
  doc_wordlist = []
  for token in doc:
    if token not in doc_wordlist:
      doc_wordlist.append(token)
      doc_word[word2id[token]] += 1
doc_word
```

```python
import math
IDF_ls = []
for i in doc_word:
  IDF_ls.append(math.log10(doc_num / i))
IDF_ls
```

```python
# TF_IDF 구하기
TF_IDF = []
for i in TF_ls:
  temp = []
  for idx, val in enumerate(i):
    temp.append(val * IDF_ls[idx])
  TF_IDF.append(temp)
TF_IDF
```
