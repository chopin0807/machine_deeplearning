```python
import requests
from bs4 import BeautifulSoup
url = "https://n.news.naver.com/mnews/article/018/0005596218?sid=101"
resp = requests.get(url)
soup = BeautifulSoup(resp.content, 'html.parser')
content = soup.find(class_='go_trans _article_content').text.replace('\n',' ')
```

## 데이터 전처리

형태소분석

```python
# 코모란
from konlpy.tag import Komoran
tokenizer = Komoran()
preprocessed_docs = []
content_token = tokenizer.morphs(content)
print(content_token)
```

한글자제거

```python
token_remove = []
for token in content_token:
  if len(token)>1:
    token_remove.append(token)
print(token_remove)
```

불용어처리

```python
stop_word = ['17','18','2023','ㄴ다','에서']
token_list = []
for token in token_remove:
  if token not in stop_word:
    token_list.append(token)
print(token_list)
```

```python
content_token = ' '.join(token_list)
content_token
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer()
tfidf_v = tfidf_vect.fit_transform([content_token])
keyword = tfidf_v.tocoo()
# coo : Coordinate 으로 0이 아닌 데이터만 별도의 배열에 저장하고, 그 데이터가 가리키는 행과 열의 위치를 별도의 배열에 저장하는 방식
```

```python
sorted_words = sorted(zip(keyword.col, keyword.data), key=lambda x:(x[1], x[0]), reverse=True)
sorted_words
```

```python
feature_name = tfidf_vect.get_feature_names_out()
feature_name
```

```python
# 핵심키워드 10개 추출
[(feature_name[i], score) for i, score in sorted_words[:10]]
```
